You will be given an array a and a value x. All you need to do is check whether the provided array contains the value.

a can contain numbers or strings. x can be either.

Return true if the array contains the value, false if not.

```python
def check(seq, elem):
    output = False
    for i in range(len(seq)):
        if(seq[i]==elem):
            output = True
            break
    return output;        

check([8, 7, 5, "bored", "of", "writing", "tests", 115], 45)
```

```python
def check(seq, elem):
    return elem in seq

check([8, 7, 5, "bored", "of", "writing", "tests", 115], 45)
```

```python
def digitize(n):
    result = [int(i) for i in str(n)]
    result.reverse() # thia reverses the list on place and returns nothing
    return result 
digitize(1234567)
```

```python
def digitize(n):
    return [int(i) for i in reversed(str(n))] # this function is a iterable function
digitize(1234567)
```

```python
def repeat_str(repeat, string):
    for _ in range(repeat):
        string.join(string)
    return string

repeat_str(5,"hello")
```

Return true if yes, false otherwise :)

('+', 4, 7) --> 11
('-', 15, 18) --> -3
('*', 5, 5) --> 25
('/', 49, 7) --> 7

```python
def basic_op(operator, value1, value2):
    #your code here
    match operator:
        case '+':
            return value1+value2
        case '-':
            return value1-value2
        case '*':
            return value1*value2
        case '/':
            return value1/value2 

basic_op('/', 4, 7)
        
```

```python
def reverse_words(text):
    word = []
    for i in text.split(' '):
        word.append(i[::-1])
    return ' '.join(word)
    
reverse_words('  double  spaced  words  ')        
```

```python
def reverse_words(text):
    return ' '.join(word[::-1] for word in text.split(' '))

reverse_words('The quick brown fox jumps over the lazy dog.')  
```
